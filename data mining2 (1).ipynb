# حالا الگوهای پرتکرار در هر شهر را بدست میاوریم
# همزمان براساس هردوگروپ بای میکنیم 

```python
grouped_df = orders.groupby(by = ['city_name_fa', 'ID_Order'])

transactionsOFcity = {}
for group, pdf in grouped_df:
    if (transactionsOFcity.get(group[0]) != None):
        transactionsOFcity.get(group[0]).append(pdf['ID_Item'].values.tolist())
        
    else:
        transactionsOFcity[group[0]] = [pdf['ID_Item'].values.tolist()]
```

```python
for eachCity in transactionsOFcity:
    patterns = pyfpgrowth.find_frequent_patterns(transactionsOFcity.get(eachCity), 2)
    rules = pyfpgrowth.generate_association_rules(patterns, 0.1)
    print("city:",eachCity,"patterns:",patterns,"rules:",rules,"\n")
```

# هیچ رولی با ایدی ادر حتی با ساپورت 2 پیدا نمیشه پس با ایدی کاستومر انجام میدیم همزمان براساس شهر و کاستومر گروپ بای میکنیم که راحتتر باشد

```python
grouped_df = orders.groupby(by = ['city_name_fa', 'ID_Customer'])

transactionsOFcity = {}
for group, pdf in grouped_df:
    if (transactionsOFcity.get(group[0])!=None):
        transactionsOFcity.get(group[0]).append(pdf['ID_Item'].values.tolist())
    else:
        transactionsOFcity[group[0]] = [pdf['ID_Item'].values.tolist()]
```

# اینجا تراکنش های یکی را در نظر نمیگیریم
# ولی میبینیم که یک رول با کانفیدنس یک و نیم میدهد که اشتباه است

```python
for eachCity in transactionsOFcity:
    transactionsFiltered2 = []
    tr=transactionsOFcity.get(eachCity)
    for transaction in tr:
        if len(transaction)>=2:
            transactionsFiltered2.append(transaction)
    patterns = pyfpgrowth.find_frequent_patterns(transactionsFiltered2,3)
    rules = pyfpgrowth.generate_association_rules(patterns,0.7)
    print("city:",eachCity,"patterns:",patterns,"rules:",rules,"\n")
```
